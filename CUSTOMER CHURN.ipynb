<h3>MODEL 1

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Extract features and target
x = customer[['tenure']]
y = customer[['Churn']]

# Split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)

# Build the model
model = Sequential()
model.add(Dense(12, input_dim=1, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=150, validation_data=(x_test, y_test), verbose=0)

# Evaluate the model
validation_accuracy = model.evaluate(x_test, y_test)[1]
print("Final Validation Accuracy:", validation_accuracy)

# Plot accuracy vs epochs
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

<H3>MODEL 2

```python
from keras.layers import Dropout
# Build the 2nd model
model_2 = Sequential()
model_2.add(Dense(12, input_dim=1, activation='relu'))
model_2.add(Dropout(0.3))
model_2.add(Dense(8, activation='relu'))
model_2.add(Dropout(0.2))
model_2.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model_2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model on the train set
history_2 = model_2.fit(x_train, y_train, epochs=150, validation_data=(x_test, y_test))
```

```python
# Print the final validation accuracy
print("Final Validation Accuracy (Model 2):", history_2.history['val_accuracy'][-1])
```

```python
# Predict values on the test set
y_pred_2 = model_2.predict(x_test).astype(int)
```

```python
# Build a confusion matrix
conf_matrix_2 = confusion_matrix(y_test, y_pred_2)
print("Confusion Matrix (Model 2):")
conf_matrix_2
```

```python
# Plot the 'Accuracy vs Epochs' graph for Model 2
plt.plot(history_2.history['accuracy'])
plt.plot(history_2.history['val_accuracy'])
plt.title('Model 2 Accuracy vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()
```

<h3>MODEL 3

```python
# Extract features and target for the 3rd model
x_3 = customer[['MonthlyCharges', 'tenure', 'TotalCharges']]
y_3 = customer[['Churn']]
```

```python

# Split the data into train and test sets
x_train_3, x_test_3, y_train_3, y_test_3 = train_test_split(x_3, y_3, test_size=0.30, random_state=42)

# Map 'No' to 0 and 'Yes' to 1
y_train_3['Churn'] = y_train_3['Churn'].map({'No': 0, 'Yes': 1})
y_test_3['Churn'] = y_test_3['Churn'].map({'No': 0, 'Yes': 1})

# Convert 'Churn' column to integer
y_train_3 = y_train_3.astype(int)
y_test_3 = y_test_3.astype(int)
```

```python
# Build the 3rd model
model_3 = Sequential()
model_3.add(Dense(12, input_dim=3, activation='relu'))
model_3.add(Dense(8, activation='relu'))
model_3.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model_3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model on the train set
history_3 = model_3.fit(x_train_3, y_train_3, epochs=150, validation_data=(x_test_3, y_test_3))
```

```python
# Print the final validation accuracy
print("Final Validation Accuracy (Model 3):", history_3.history['val_accuracy'][-1])
```

```python
# Predict values on the test set
y_pred_3 = model_3.predict(x_test_3).astype(int)
```

```python
# Build a confusion matrix
conf_matrix_3 = confusion_matrix(y_test_3, y_pred_3)
print("Confusion Matrix (Model 3):")
conf_matrix_3
```

```python
# Plot the 'Accuracy vs Epochs' graph for Model 3
plt.plot(history_3.history['accuracy'])
plt.plot(history_3.history['val_accuracy'])
plt.title('Model 3 Accuracy vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()
```
